Given a $\mathbb{R}=2$ vector in a 3-dimensional plane and visualize it using the cells below. In the first markdown cell below, write your vector using LaTeX. You should  represent it as a system of linear equations. In your methodology discuss the different functions that you have used, the equation of your vectors, and the flowchart for declaring and displaying your vector. Please make sure that your flowchart has only few words and not putting the entire code as it is bad practice. In your results, display and discuss the vector visualization code you made.

$$A =  \left\{
    \begin{array}\\
    2x , 2y , 3z\\
    3x , 3y , 2z\\
    2x , 2y , 6z
     \end{array}
\right.
$$

```python
A = np.array([
    [2,5,3],
    [3,3,2],
    [2,2,6],
])

fig = plt.figure()
ax1 = fig.gca(projection='3d')

ax1.set_xlim([0, 6])
ax1.set_ylim([0, 6])
ax1.set_zlim([0, 6])

origin = (0,0,0)
ax1.quiver(origin, origin, origin, A[:,0], A[:,1], A[:,2], 
           arrow_length_ratio=0.05, colors=['red','blue','green'])
plt.show()
```
